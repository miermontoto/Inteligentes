## **Distribuciones**

Las distribuciones se obtienen con el objetivo de mostrar los posibles valores que puede poseer una variable concreta. Además podemos ver la frecuencia de los mismos.

_Para una variable:_
* Histograma : `sns.histplot` o `sns.countplot`
* KDE (Kernel Density Estimate): `sns.kdeplot`

_Para varias variables:_
* KDE2D: `sns.jointplot`

Histograma de la columna "age" (variable continua).

````
plt.figure(figsize=fig_size); # Tamaño del plot
histo = sns.histplot(data=data['age'], kde=False, bins=range(15,46));
#histo.set_xticks(range(15,46)); # Etiquetas del eje X
histo.set_xlabel("Edad")
histo.set_ylabel("Número de ejemplos")
histo.set_title("Histograma de la variable 'age'")
````

Histograma de la columna "preferred_foot" (variable categórica).

````
plt.figure(figsize=fig_size); # Tamaño del plot
histo = sns.countplot(x=data['preferred_foot']);
histo.set_xlabel("Pie preferido")
histo.set_ylabel("Número de ejemplos")
histo.set_title("Histograma de la variable 'preferred_foot'")
````

KDE de la columna "age".


````
plt.figure(figsize=fig_size);
kde_plot = sns.kdeplot(data=data['age'], shade=True);
kde_plot.set_xticks(range(15,46)); # Etiquetas del eje X
kde_plot.set_xlabel("Edad")
kde_plot.set_ylabel("Densidad")
kde_plot.set_title("KDE de la variable 'age'")
````

KDE de la columna "age" de 1 equipo frente a otro.

````
plt.figure(figsize=fig_size);

plt_data = data.loc[data["club"].isin(["Real Madrid CF","FC Barcelona"])];
kde_plot = sns.kdeplot(data=plt_data, x="age", hue="club", shade=True);

kde_plot.set_xticks(range(15,46)); # Etiquetas del eje X
kde_plot.set_xlabel("Edad")
kde_plot.set_ylabel("Densidad")
kde_plot.set_title("KDE de la variable 'age'")
````

KDE2D de la columna "age" y de la columna "overall".

````
plt.figure(figsize=fig_size);
kde_plot = sns.jointplot(x=data['age'], y=data['overall'], kind="kde");
````

> **Obtén la distribución de la variable "weight_kg", la de "height_cm" y la de ambas juntas**


````
plt.figure(figsize=fig_size);
weight_plot = sns.kdeplot(data=data['weight_kg'], shade=True)
weight_plot.set_xlabel('Peso (kg)')

height_plot = sns.kdeplot(data=data['height_cm'], shade=True)
height_plot.set_xlabel('Altura (cm)')

both_plot = sns.jointplot(x=data['weight_kg'], y=data['height_cm'], kind='kde')
````

````
from sklearn.decomposition import PCA # Importar PCA
pca = PCA(n_components=2, random_state=10) # Configuración para reducir a 2 dimensiones y fijar la semilla.

plot_data = data.sample(2000, random_state=10) # Seleccionar una muestra del conjunto (2000 elementos)
plot_data = plot_data[["weight_kg","height_cm","age","overall"]] # Seleccionar las columnas que se quieren representar
new_data = pca.fit_transform(plot_data.values) # Transformar nuestros datos a 2 dimensiones

plt.figure(figsize=fig_size); # Tamaño del plot
the_plot = sns.scatterplot(x=new_data[:,0], y=new_data[:,1]); # Hacer gráfico de dispersión
the_plot.set_xlabel("Nueva dimensión X")
the_plot.set_ylabel("Nueva dimensión Y")
the_plot.set_title("Reducción de dimensiones mediante PCA")
````

````
from sklearn.manifold import TSNE # Importar tSNE
tsne = TSNE(n_components=2, random_state=10) # Configuración para reducir a 2 dimensiones y fijar la semilla.

plot_data = data.sample(2000, random_state=10) # Seleccionar una muestra del conjunto (2000 elementos)
plot_data = plot_data[["weight_kg","height_cm","age","overall"]] # Seleccionar las columnas que se quieren representar
new_data = tsne.fit_transform(plot_data.values)# Transformar nuestros datos a 2 dimensiones

plt.figure(figsize=fig_size); # Tamaño del plot
the_plot = sns.scatterplot(x=new_data[:,0], y=new_data[:,1]); # Hacer gráfico de dispersión
the_plot.set_xlabel("Nueva dimensión X")
the_plot.set_ylabel("Nueva dimensión Y")
the_plot.set_title("Reducción de dimensiones mediante t-SNE")
````

# **4. Visualización dinámica**
---

En este último apartado haremos uso de la librería Bokeh.  
Esta librería genera código html en vez de generar una imágen estática.

MNIST será el conjunto que utilizaremos en este caso.  
Este conjunto está formado por 70000 imágenes en blanco y negro de tamaño 28*28.
Cada una de ellas representa un dígito del 0 al 9.  

Esta vez obtendremos el conjunto utilizando una librería que se encarga de descargarlo.

````
from sklearn.datasets import fetch_openml # Cargamos la librería que contiene los datos
mnist = fetch_openml('mnist_784', cache=False) # Descargamos el conjunto MNIST
````

 En este caso no tenemos un DataFrame, tenemos un diccionario.

````
mnist.keys() # Obtenemos las claves
````

Los datos relativos a las imágenes se encuentran en la clave "data".

````
images = mnist.data.astype('float32'); # Cada una de las 70000 imágenes se encuentran codificadas en un vector de 784 dimensiones (28*28)
images.shape # Verificamos el tamaño
````

Otro dato relevante es el dígito que representa cada imagen. La clave "target" tiene esta información.

````
digit = mnist.target.astype('int64')
digit # Vemos el contenido
````

Vamos a obtener un subconjunto más pequeño para poder hacer pruebas de forma sencilla.

````
sub_images = images[:2000] # Nos quedamos con las 2000 primeras imágenes
sub_digit = digit[:2000] # Nos quedamos con los 2000 primeros dígitos
````

Como cada uno de nuestros ejemplos está representado por más de una variable (784 exactamente) necesitaremos hacer una reducción de dimensionalidad (PCA o tSNE).

````
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=100, verbose=1) # Reducimos a 2 dimensiones con tSNE
plot_data = tsne.fit_transform(sub_images) # Introducimos los datos
````

Cargamos la librería Bokeh.

````
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import output_notebook

output_notebook() # Configurar Bokeh
````

Asignar a cada dígito un color.
